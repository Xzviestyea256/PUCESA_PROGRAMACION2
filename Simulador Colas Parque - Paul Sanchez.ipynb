```python
from collections import deque
import time

#Creación de las 3 colas de prioridad
cola_general = deque()
cola_prioritaria = deque()
cola_discapacidad = deque()
#Creación de variables que serviran más luego, al momento de generar las estadísticas
#Cuentan la cantidad de pasajeron que suben de cada cola de prioridad, y de manera general
conticket = 0
contdis = 0
contprip = 0
contgen = 0
viajes_totales = 0
#vijes_totales solo cuenta los viajes realizados, ya que el sistema no realiza un viaje si no hya almenos 4 
#pasajeros para poner en un carrito

def menu():
    #Función que muestra el menú al usuario
    while True:
        print(f"\n---Menú de Sistema del Parque---")
        print("1. Agregar perosna")
        print("2. Ver estado de colas")
        print("3. Simular viaje")
        print("4. Salir del sistema")
        
        #Por lo visto para cuando se tiene que ingresar datos mediante un input se abre arriba una pestañita para ingresar todo
        #asi que pilas por ahi, eso recien me di cuenta hoy 
        seleccion = int(input("Seleccione una opcion (1-4)"))
        print(f"Selección = {seleccion}")
        if seleccion == 1:
            #Llama a la funcion de ingreso_manual para que el propio usuario ingrese a los pasajeros y su piroirdad
            ingreso_manual()
        if seleccion == 2:
            #Llama a la funcion mostrar_colas para ver el estado de las 3 colas
            print("---Estado de las colas de pasajeros---")
            mostrar_colas()
        if seleccion == 3:
            #Llama a la fucnión iniciar_simulacion para que se ejecute la simulación
            turnos_totales = int(input("Ingrese la cantidad de turnos totales: "))
            iniciar_simulacion(turnos_totales)
        if seleccion == 4:
            #Llama a la funcion estadisticas, que muestra las estadisticas finales de todos los viajes realizados
            #y despues cierra el programa
            estadisticas()
            break

            
def ingreso_manual():
    #Funcion que permite al usuario ingresar los daots (nombre y prioridad) del pasajero
    while True:
        print("---Ingreso de pasajeros---")
        nombre = input("Ingrese el nombre del pasajero: ")
        prioridad = int(input("Ingrese la prioridad del pasajero (1(general)-2(prioritaria)-3(discapacidad)): "))
        print(f"Ingresa: {nombre}, Prioridad: {prioridad}")
        agregar_visitante(nombre, prioridad)
        continuar = int(input("Desea seguir ingresando pasajeros? 1(si)/2(no):"))
        if continuar == 1:
            continue
        if continuar == 2:
            break


def agregar_visitante(nombre, prioridad):
    #Funcion que ubica a un pasajero enuna de las 3 colas segun su prioridad
    if prioridad == 1:
        cola_general.append(nombre)
    elif prioridad == 2:
        cola_prioritaria.append(nombre)
    elif prioridad == 3:
        cola_discapacidad.append(nombre)

def mostrar_colas():
    #Funcion que muestra el estado de todas las colas
    print("\nCola Discapacidad:", list(cola_discapacidad))
    print("Cola Prioritaria:", list(cola_prioritaria))
    print("Cola General:", list(cola_general))
    
CAPACIDAD_VAGON = 4

def cargar_vagon():
    #Funcion que ubica a 4 pasajeros dentro del vagon para el viaje
    #Siempre agregara primero a los de la cola de discapacidad, luego a los de la cola prioritaria
    #y al final a los de la cola general
    pasajeros = []

    while len(pasajeros) < CAPACIDAD_VAGON:
        if cola_discapacidad:
            global conticket
            conticket += 1
            global contdis
            contdis += 1
            #Los contadores llevan la cuenta (valga la redundancia) de cuantos pasajeros ingresan 
            #al carrito, de manera general, y de por cada cola de prioridad, asi se peuden generar
            #las estadisticas de manera correcta
            ticket(cola_discapacidad[0], "Discapacidad", viaje_num)
            #Despues de agregar al primer -pasajero de la cola al vagon, lo elimina de su respectiva cola.
            pasajeros.append(cola_discapacidad.popleft())
        elif cola_prioritaria:
            conticket += 1
            global contprip
            contprip += 1
            ticket(cola_prioritaria[0], "Prioritaria", viaje_num)
            pasajeros.append(cola_prioritaria.popleft())
        elif cola_general:
            conticket += 1
            global contgen
            contgen += 1
            ticket(cola_general[0], "General", viaje_num)
            pasajeros.append(cola_general.popleft())
        else:
            break
    return pasajeros

DURACION_VIAJE = 3

def simular_viaje(pasajeros, numero_viaje):
    global viajes_totales
    viajes_totales += 1
    print(f"\n Viaje #{numero_viaje} con: {pasajeros}")
    #Pregunta al usuario cuanto tiempo durará cada viaje
    duracion = int(input("Ingrese la duración del viaje: "))
    for t in range(DURACION_VIAJE):
        print(f"  Turno {t+1}/{DURACION_VIAJE}")
        time.sleep(duracion)
    print("¡Viaje finalizado!")

def iniciar_simulacion(turnos_totales):
    #Funcion que se asegura de que el viaje es posible (minimo 4 pasajeros)
    global viaje_num
    viaje_num = 1
    
    for turno in range(turnos_totales):
        print(f"\n--- Turno {turno+1}/{turnos_totales} ---")

        if len(cola_general) + len(cola_prioritaria) + len(cola_discapacidad) >= CAPACIDAD_VAGON:
            pasajeros = cargar_vagon()         
            #Llama a la funcion simular_viaje para que el viaje se realize de manera correcta
            simular_viaje(pasajeros, viaje_num) 
            #Le suma 1 al valor del viaje que realiza el carrito (en este caso entiendo 1ue son las vetas que
            # le da al recorrido)
            viaje_num += 1                   
        else:
            print("Aún no hay suficientes personas para el viaje.")

def ticket(nombre, prioridad, viaje_num):
    #Genera un ticket con los datos del pasajero que sube al vagón (nombre, prioridad y el numero del viaje al que sube)
    print(f"Ticket: {nombre} | Prioridad: {prioridad} | Num_viaje: {viaje_num}")
    print("Gracias por visitar el parque")
    
def estadisticas():
    #Funcion que muestra las estadisticas totales de todos los recorridos realizados durante la ejecucuín del programa
    print("\n------Estadísticas------")
    #Muestra el numero total de pasajeros que subieron al carrito
    print(f"Total de tickets emitidos: {conticket}")
    #Muestra la cantidad de pasajeros de cada prioridad (discaacidad, prioritaria y genera) que subieron al carrito
    print(f"Tickets Discapacidad: {contdis}")
    print(f"Tickets Prioritarios: {contprip}")
    print(f"Tickets Generales: {contgen}")
    #Muestra la cantidad total de viajes que se realizaron
    print(f"Total de viajes realizados: {viajes_totales}")
    #Muestra un promedio entre la cantidad de pasajeros que subieron al carrito y los viajes realizados -
    print(f"Promedio de personas por viaje: {conticket // viajes_totales}")
    

menu()
```
